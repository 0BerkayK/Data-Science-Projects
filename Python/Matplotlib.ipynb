```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ageList = [10,20,30,30,40,40,50,50,60,70]
```

```python
weightList = [20,60,70,75,75,80,85,85,90,95]
```

```python
numpyAgeList = np.array(ageList)
numpyWeightList = np.array(weightList)
```

```python
numpyAgeList
```

```python
numpyWeightList
```

```python
plt.plot(numpyAgeList,numpyWeightList,"r")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Weight change chart according to age")
## plt.show()
```

```python
numpyArray1 = np.linspace(0,10,20)
```

```python
numpyArray1
```

numpyArray2 = numpyArray1 ** 3
numpyArray2

```python
plt.plot(numpyArray1,numpyArray2, "b*-")
```

```python
plt.subplot(1,2,1) # first => row , second => column , third => number 
plt.plot(numpyArray1,numpyArray2, "r*-")
plt.subplot(1,2,2) 
plt.plot(numpyArray2,numpyArray1, "b--" )
```

```python
myFigure = plt.figure()
figureAxes = myFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyArray1,numpyArray2)
figureAxes.set_xlabel("X axis")
figureAxes.set_ylabel("Y axis")
figureAxes.set_title("Graph")
```

```python
figure2 = plt.figure()

axis1 = figure2.add_axes([0.3,0.3,0.7,0.7])
axis2 = figure2.add_axes([0.2,0.5,0.5,0.5])

axis1.plot(numpyArray1,numpyArray2, "g")

axis1.set_xlabel("X axis")
axis1.set_ylabel("Y axis")
axis1.set_title("Graph")

axis2.plot(numpyArray2,numpyArray1, "b")

axis2.set_xlabel("X axis")
axis2.set_ylabel("Y axis")
axis2.set_title("Graph")

```
